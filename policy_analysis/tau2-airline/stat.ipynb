```python
import json

file = "spec.json"

with open(file, "r") as f:
    data = json.load(f)
```

```python
tag_count = {}

for item in data:
    tag = item.get("tag", "no_tag")
    if tag in tag_count:
        tag_count[tag] += 1
    else:
        tag_count[tag] = 1
print(tag_count)
```

```python
guardrail_count = {}

for item in data:
    sym_rail = item.get("sym guardrail", [])
    if not sym_rail:
        continue
    # sort the list to ensure consistent counting
    sym_rail.sort()
    sym_rail_str = ", ".join(sym_rail)
    if sym_rail_str in guardrail_count:
        guardrail_count[sym_rail_str] += 1
    else:
        guardrail_count[sym_rail_str] = 1
print(guardrail_count)
```
