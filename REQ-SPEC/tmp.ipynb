````python
import json

file = "spec.json"

with open(file, "r") as f:
    data = json.load(f)
````

````python
tag_count = {}

for item in data:
    tag = item.get("tag", "no_tag")
    if tag in tag_count:
        tag_count[tag] += 1
    else:
        tag_count[tag] = 1
print(tag_count)
````

````python
template_type_count = {}

for item in data:
    if item.get("tag") == "enforcable":
        for spec in item.get("spec", []):
            template_type = spec.get("template_type", "no_template_type")
            if template_type in template_type_count:
                template_type_count[template_type] += 1
            else:
                template_type_count[template_type] = 1
print(template_type_count)
````
