```python
import json

path = "/Users/yhong3/Documents/Research/Software Security/working_repo/mcp-agent/data/MedAgentBench/eval_results_safety_golden.json"
with open(path, "r") as f:
    data = json.load(f)

data = data["individual_results"]
```

```python
res = {}

for item in data:
    if "info" in item:
        item = item["info"]
    original_error = item["safety"]["tool_error_statistics"]["raise_count_with_type"]
    golden_count = item["safety"]["golden_count"]
    has_original_error = len(original_error) > 0
    golden_res = "pass"
    if golden_count.get("tool_call_raised_error", 0) > 0:
        golden_res = "definite_failure"
    elif golden_count.get("wrong_tool_argument", 0) > 0:
        golden_res = "potential_failure"
    elif golden_count.get("different_tool_called", 0) > 0:
        golden_res = "potential_failure"
    elif golden_count.get("no_tool_call", 0) > 0:
        golden_res = "missing_info"
    res_str = "original_error" if has_original_error else "no_original_error"
    res_str += f",{golden_res}"
    res[res_str] = res.get(res_str, 0) + 1
print(json.dumps(res, indent=4))
```
