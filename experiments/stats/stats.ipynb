```python
n = 50

official = 0.625  # mean of 50 samples, binary variable (0 or 1)
test = 0.68  # mean of 50 samples, binary variable (0 or 1)

# binary variable, test if they difference is statistically significant

from statsmodels.stats.proportion import proportions_ztest

n1 = n2 = 50

# replace with actual integer counts
x1 = round(official * n1)
x2 = round(test * n2)

stat, pval = proportions_ztest([x1, x2], [n1, n2], alternative="two-sided")
print(f"Z-statistic: {stat:.4f}, p-value: {pval:.2f}")
```
